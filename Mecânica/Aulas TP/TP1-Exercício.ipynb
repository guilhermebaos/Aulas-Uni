Isto aqui são as minhas equações preferidas.

$$

\\ \ \\

\\ \ \\
1 + 1 = 2
$$

Não te esqueças de espreitar o [SimulaFQ](https://www.simulafq.pt)!

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
tt = np.linspace(-2, 2, 1001)
```

```python
xx = math.e ** (-1 * xx ** 2)
```

```python
print('Tabela de Posições')
print('-'*20)
print('t/s      x/m')
for t, x in zip(tt, xx):
    print(f'{t:.3f}   {x:.3f}')
```

```python
plt.plot(tt, xx, 'r', label='$x(t)$')
plt.xlabel('Tempo/ s')
plt.ylabel('Posição/ m')
plt.grid(True)
plt.legend()
```

# Problema de Fermi
Vamos tentar estimar o número de grãos de arroz num saco de 1kg

## Estimativa

```python
# Vamos supor que o grão de arroz é aproximadamente igual a um cilindro
raio = 0.5e-03
altura = 1e-02
volume_grao = math.pi * raio ** 2 * altura
```

```python
# Vamos supor que o saco de arroz de 1kg é aproximadamente um paralelepípedo
largura= 15e-02
altura = 20e-02
profundidade = 10e-02
espaco_usado = 0.8
volume_saco = largura * altura * profundidade * espaco_usado
```

```python
# Vamos ver quantos grão de arroz há num saco de 1kg
n = (volume_saco / volume_grao)
int(n)
```

## Confirmação e Fator de Erro

Peso de um grão de arroz, tirado da Internet: $\frac{1}{64}$ g.

Número de grão de arroz num saco de 1kg:

```python
# Número real de grão de arroz num saco
n_real = (1 / (1e-03/64))
int(n_real)
```

```python
# Fator de Erro
print(max([n_real / n, n / n_real]))
```
