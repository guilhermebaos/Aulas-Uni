## Equação de onda com coeficientes variáveis

Se o meio não fôr homogéneo, a velocidade pode variar de ponto para ponto. Teremos então uma equação da forma:

$$
\frac{\partial^2 u}{\partial t^2} =
\frac{\partial}{\partial x}\left( q(x)
\frac{\partial u}{\partial x}\right) + f(x,t)
$$

que será amostrada nos pontos $(x_i,t_n)$.

### Discretizar o coeficiente variável

A ideia é começar por discretizar a derivada exterior. Definindo

$$\phi=q(x)\frac{\partial u}{\partial x},$$

e usando derivadas centrais em torno de $x=x_i$ para a derivada de $\phi$:
$$
\left[\frac{\partial \phi}{\partial x}\right]^n_i \approx \frac{\phi_{i+\frac{1}{2}}−\phi_{i-\frac{1}{2}}}{\Delta x}=[\delta_x \phi]^n_i.
$$

Depois discretizamos os operadores internos:
$$
\phi_{i+\frac{1}{2}}=q_{i+\frac{1}{2}}\left[\frac{\partial u}{\partial x}\right]^n_{i+\frac{1}{2}}\approx q_{i+\frac{1}{2}}  \frac{u^n_{i+1}−u^n_i}{\Delta x}=[q \delta_x u]^n_{i+\frac{1}{2}}
$$

De um modo análogo,
$$
\phi_{i-\frac{1}{2}}=q_{i-\frac{1}{2}}\left[\frac{\partial u}{\partial x}\right]^n_{i-\frac{1}{2}} \approx q_{i-\frac{1}{2}} \frac{u^n_i−u^n_{i−1}}{\Delta x}=[q \delta_x u]^n_{i-\frac{1}{2}}.
$$

Combinando estes resultados podemos escrever:

$$
\left[
\frac{\partial}{\partial x}\left( q(x)
\frac{\partial u}{\partial x}\right)\right]^n_i
\approx \frac{1}{\Delta x^2}
\left( q_{i+\frac{1}{2}} \left({u^n_{i+1} - u^n_{i}}\right)
- q_{i-\frac{1}{2}} \left({u^n_{i} - u^n_{i-1}}\right)\right)
$$

ou em notação de operador:

$$
\left[
\frac{\partial}{\partial x}\left( q(x)
\frac{\partial u}{\partial x}\right)\right]^n_i
\approx [\delta_xq \delta_x u]^n_i
$$

As expressões anteriores envolvem o cálculo de $q^n$ em pontos intermédios da grelha espacial. Para isso podemos usar uma média entre os pontos vizinhos, que pode tomar várias formas:

$$
\begin{align*}
q_{i+\frac{1}{2}} &\approx
\frac{1}{2}\left( q_{i} + q_{i+1}\right) =
[{\overline{q}}^{x}]_i
\quad &\hbox{(média aritmética)}
\\ 
q_{i+\frac{1}{2}} &\approx
2\left( \frac{1}{q_{i}} + \frac{1}{q_{i+1}}\right)^{-1}
\quad &\hbox{(média harmónica)}
\\ 
q_{i+\frac{1}{2}} &\approx
\left(q_{i}q_{i+1}\right)^{1/2}
\quad &\hbox{(média geométrica)},
\end{align*}
$$

sendo a primeira a mais usada. Com essa escolha, e resolvendo e ordem a $u^{n+1}_1$, teremos:

$$
\begin{align}
u^{n+1}_i &= - u_i^{n-1}  + 2u_i^n + \left(\frac{\Delta t}{\Delta x}\right)^2\times \nonumber\\ 
&\quad  \left(
\frac{1}{2}(q_{i} + q_{i+1})(u_{i+1}^n - u_{i}^n) -
\frac{1}{2}(q_{i} + q_{i-1})(u_{i}^n - u_{i-1}^n)\right)
+ \nonumber\\ 
& \quad \Delta t^2 f^n_i
\end{align}
$$

### Condições fronteira de von Neumann para coeficientes variáveis

Considerando como condição fronteira $\partial u/\partial x=0$ em $x=L=N_x\Delta x$ virá:

$$
\frac{u_{i+1}^{n} - u_{i-1}^n}{2\Delta x} = 0\quad u_{i+1}^n = u_{i-1}^n,
\quad i=N_x.
$$

Usando este resultado no stencil geral calculado para $i=N_x$, temos:

$$
u^{n+1}_i \approx
- u_i^{n-1}  + 2u_i^n + \left(\frac{\Delta t}{\Delta x}\right)^2
2q_{i}(u_{i-1}^n - u_{i}^n) + \Delta t^2 f^n_i
$$

(usando $q_{i+\frac{1}{2}} + q_{i-\frac{1}{2}}\approx 2q_i$, mas também podíamos ter usado $dq/dx=0$).

## Caso ainda mais geral

$$
\varrho(x)\frac{\partial^2 u}{\partial t^2} =
\frac{\partial}{\partial x}\left( q(x)
\frac{\partial u}{\partial x}\right) + f(x,t)
$$

Podemos discretizar usando a equação em operadores:

$$
[\varrho D_tD_t u = D_x\overline{q}^xD_x u + f]^n_i,
$$

em que $\varrho$ é calculado no próprio ponto.

O caso mais geral a tratar tem então a forma:

$
\begin{align}
\varrho(x)u_{tt} &= (c^2(x)u_x)_x + f(x,t),\quad &x\in (0,L),\ t\in [0,T]
\\ 
u(x,0) &= I(x),\quad &x\in [0,L]
\\ 
u_t(x,0) &= V(x),\quad &x\in [0,L]
\\ 
u(0,t) &= U_0(t)\hbox{ ou } u_x(0,t)=0,\quad &t\in[0,T]
\\ 
u(L,t) &= U_L(t)\hbox{ ou } u_x(L,t)=0,\quad &t\in [0,T]
\end{align}
$
