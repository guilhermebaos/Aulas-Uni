## Equação de Schödinger dependente do tempo

A equação de Schrödinger dependente do tempo tem a forma de uma equação de reacção-difusão:

$$
i\frac{\partial\psi}{\partial t} = -\nabla^2\psi + V\psi  \tag{1} 
$$

Notando que pode ser reescrita como:

$$
\frac{\partial\psi}{-i\partial t} = -\nabla^2\psi + V\psi \tag{1a}
$$

vemos que é como a equação de difusão que estudamos antes, mas com um tempo imaginário!

Usando o método de Crank-Nicholson, fazemos:

$$
i\frac{\psi^{n+1}_j-\psi_j}{\Delta t} = -\frac{1}{2}\left(\frac{\psi_{j+1}^{n+1}-2\psi_j^{n+1}+\psi_{j-1}^{n+1}}{\Delta x^2}+\frac{\psi_{j+1}^{n}-2\psi_j^{n}+\psi_{j-1}^{n}}{\Delta x^2}\right)  +\frac{1}{2}\left(V_j\psi_j^{n+1} +V_j\psi_j^n\right)
$$

onde se supôs que $V = V(x)$, independente do tempo. (Se $V=V(x,t)$ teremos que separar a contribuição no tempo presente e no tempo futuro). Isto leva ao sistema de equações:

$
\begin{align*}
\frac{1}{2}\frac{\Delta t}{\Delta x^2}\psi_{j+1}^{n+1} & +\left(i-\frac{\Delta t}{\Delta x^2}-\frac{1}{2}\Delta t V_j\right)\psi_j^{n+1}+\frac{1}{2}\frac{\Delta t}{\Delta x^2}\psi_{j-1}^{n+1}=\\
&-\frac{1}{2}\frac{\Delta t}{\Delta x^2}\psi_{j-1}^n + \left(i+\frac{\Delta t}{\Delta x^2}+\frac{1}{2}\Delta t V_j\right)\psi_j^{n} -\frac{1}{2}\frac{\Delta t}{\Delta x^2}\psi_{j+1}^n\tag{2}
\end{align*}
$

que tem a forma:

$\left(\begin{array}{ccccc}A_0 & A_- & 0 & 0 &\cdots\\ 
                         A_+ & A_0 & A_- & 0 &\cdots\\
                         0 & A_+ & A_0 & A_- &\cdots \\
                         \vdots & \vdots & \vdots & \ddots &  \vdots\end{array}\right)\left(\begin{array}{c}\psi_0^{n+1} \\ \psi_1^{n+1}\\ \vdots \\ \psi_{J-1}^{n+1}\end{array}\right)  =  \left(\begin{array}{ccccc}B_0 & B_- & 0 & 0 &\cdots\\ 
                         B_+ & B_0 & B_- & 0 &\cdots\\
                         0 & B_+ & B_0 & B_- &\cdots \\
                         \vdots & \vdots & \vdots & \ddots &  \vdots\end{array}\right) \left(\begin{array}{c}\psi_0^{n} \\ \psi_1^{n}\\ \vdots \\ \psi_{J-1}^{n}\end{array}\right)
$

onde $A_{+} = A_{-} = \alpha \equiv \frac{\Delta t}{2 \Delta x^2}$,  $A_0 = i - 2\alpha - \frac{\Delta t V_{j}}{2}$, $B_{+} = B_{-} = -\alpha$, e $B_0 = i + 2\alpha + \frac{\Delta t V_j}{2}$.

É claro que podíamos utilizar os métodos explícito ou implícito, com as limitações vistas antes. Mas aqui surge um problema adicional. Como sabemos, a função de onda mede uma probabilidade, e como tal tem uma relação de normalização. Ora pode-se mostrar que nem o método explícito, nem o implícito, conservam a norma! Já o método de CN é unitário, ié, conserva a norma a cada passo.

### Estabilidade
De onde vem a não-unicidade?

Resolver a equação de Schödinger dependente do tempo (ESDT) corresponde a saber como calcular:

$
\left| \psi(x,t) \right\rangle = e^{-iHt}\left|\psi(x,0) \right \rangle,
$

que contém um operador diferencial na exponencial.

Se aplicarmos diferenças finitas para a frente, o operador diferencial tem a forma de uma matriz tridiagonal, e a exponencial pode ser aproximada pelos dois primeiros termos da série de Taylor:

$
e^{-iHt} \approx 1 - iHt
$

Esta discretização e linearização é o que dá origem à não-unicidade.

De um modo análogo se mostra que a discretização para trás leva a:

$
e^{-iHt} \approx 1 + iHt
$


que também não é unitária.


Já a combinação da discretização para frente (explícito) com a combinação para trás (implícito) leva à aproximação:

$
e^{-iHt} \approx  \frac{1-\frac{1}{2}iHt}{1+\frac{1}{2}iHt}
$

que facilmente se vê que é unitária.

### Métodos alternativos
Existem outros métodos, como por exemplo o método apresentado por Vesscher (ver ref), que separa as partes real e imaginária da função de onda, $\psi =R + iI$, e avança cada uma delas alternadamente a cada meio passo do tempo:

$\begin{eqnarray}\frac{dR}{dt}&=&HI \\
\frac{dI}{dt}&=&-HR \\
H&=&-\frac{1}{2m}\nabla^2+V\end{eqnarray}
$

Então $R$ é calculado para todos os "passos" ($t = 0, \Delta t, 2\Delta t. \ldots$) e $I$ é calculado para todos os "meios-passos" ($t = \Delta t/2, 3\Delta t/2. \ldots$) como:

$
R(t+\frac{1}{2} \Delta t)=R(t-\frac{1}{2} \Delta t)+\Delta t HI(t)
$

$
I(t+\frac{1}{2} \Delta t)=I(t-\frac{1}{2} \Delta t)-\Delta t HR(t)
$

onde

$
\nabla^2\psi(r,t)=\frac{\psi(r+\Delta r,t)-2\psi(r,t)+\psi(r-\Delta r,t)}{\Delta r^2}
$

A densidade de probabilidade pode ser definida como: the probability density as

$
P(x,t)=R^2(x,t)+I(x,t+\frac{1}{2}\Delta t)I(x,t-\frac{1}{2}\Delta t)
$

em passos inteiros no tempo, e

$
P(x,t)=R(x,t+\frac{1}{2}\Delta t)R(x,t-\frac{1}{2}\Delta t)+I^2(x,t)
$

em passos semi-inteiros no tempo, o que faz o algoritmo unitário, conservando assim a probabilidade. 

http://www.hunter.cuny.edu/physics/courses/physics485/repository/files/reference-papers/Visscher%20NumSol%20Schrodinger%20Eqt.pdf

Outro método, que veremos mais adiante, é o método de _split-step Fourier_, um método pseudo-espectral. Nele separamos (_split_) os dois operadores, o da energia cinética ("difusão"), e o do potencial (local). Cada um é aplicado alternadamente, sendo que o operador diferencial tem um tratamento exacto no espaço de Fourier.
