```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Formula: https://en.wikipedia.org/wiki/Lennard-Jones_potential
eps = 1
sigma = 1


# Create Values
rr = np.linspace(0.8, 2 * sigma, 500)
LJ =  4 * eps * ((sigma / rr)**12 - (sigma / rr)**6)

LJd =  4 * eps * (12 * sigma**12 * rr**(-13) - 6 * sigma ** 6 * rr**(-7))
epsd = abs(min(LJd))


# Plot
fig, ax1 = plt.subplots()
plt.title("Lennard-Jones Potential")

# Primary Axis
ax1.plot(rr, LJ, "b", label="Potential")
ax1.set_ylabel('Energy')
ax1.set_ylim(-eps * 1.1, 2 * eps)

# Secondary Axis
ax2 = ax1.twinx()
ax2.plot(rr, LJd, "r", label="Force")
ax2.set_ylabel('Force')
ax2.set_ylim(-epsd * 1.1, 2 * epsd)

# Show
plt.xlabel("Distance z")
fig.legend()
plt.grid()
```
