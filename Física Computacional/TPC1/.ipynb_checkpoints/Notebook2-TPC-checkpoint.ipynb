# 1. Série de Fourier

Uma função diz-se periódica se para todos os valores de $x$ verifica-se que $f(x+L)=f(x)$. Estas funções podem ser expressas à custa de uma base de funções periódicas, 
\begin{align}
f(x)=\sum_{n=-\infty}^{+\infty}c_{n}e^{ik_{n}x}
\end{align}
com os valores $k_{n}=2\pi n/L$ e os coeficientes $c_{n}$ dados por 
\begin{align}
c_{n}=\frac{1}{L}\int_{0}^{L}dxf(x)e^{-ik_{n}x}.
\end{align}
Esta expansão chama-se uma série de Fourier e pode ser entendida como decomposição numa base infinita de funções designadas por,
\begin{align}
\phi_{n}(x)=e^{ik_{n}x}
\end{align}
e na qual podemos definir um produto interno entre duas funções $f(x)$ e $g(x)$ como,
\begin{align}
f\cdot g=\frac{1}{L}\int_{0}^{L}dxf^{*}(x)g(x).
\end{align}
Com esta interpretação os coeficientes $c_{n}$ podem ser compreendidos como as projecções de uma função em cada elemento da base. 

1. Mostre analiticamente a ortogonalidade dos elementos da base, ou seja $\phi_{n}\cdot\phi_{m}=\delta_{mn}$.

#Responda Aqui

A demonstração feita para as bases foi:

- Quando temos m = n:
- $e^{0} = 1$

$$ \phi_n . \phi_m = \frac{1}{L} \int_{0}^{L}  e^{-i k_n x}  e^{i k_m x} dx$$

$$ \equiv \frac{1}{L} \int_{0}^{L}  e^{\frac{2 \pi i x}{L} (-n + m)}  dx  =  \frac{1}{L} \int_{0}^{L}  e^{\frac{2 \pi i x}{L} 0}  dx = \frac{1}{L} \int_{0}^{L}  e^{0}  dx$$    
  
$$ \frac{1}{L} \int_{0}^{L} dx =   \frac{L}{L} = 1$$


- Quando m $\neq$ n:

$$\phi_n . \phi_m = \frac{1}{L} \int_{0}^{L}  e^{-i k_n x}  e^{i k_m x} dx$$

$$ = \frac{1}{L} \int_{0}^{L}  e^{\frac{i2\pi x}{L}(m-n)} dx  = \frac{1}{L} \left[\frac{L}{2\pi i(m-n)} e^{\frac{i2\pi x(m-n)}{L}}\right]_{0}^L = \frac{1}{L} \left[\frac{L}{2\pi i(m-n)} e^{i2\pi (m-n)} - \frac{L}{2\pi i(m-n)}\right]$$



$$= \frac{1}{2\pi i(m-n)}(e^{i2\pi (m-n)} - 1) \ \ \forall  m,n \in \mathbb{Z} : e^{2\pi i } = 1$$

$$\equiv \frac{1}{2\pi i(m-n)}(1-1) = 0$$

3. Construa uma função que calcule uma série de coeficientes de Fourier para uma função arbitrária usando o método de Simpson. A função deve receber como argumentos pelo menos o parâmetro $L$, um array de $k_{n}$ e o número de intervalos de integração. 

#Responda Aqui

Ao criarmos uma função que recebe como argumentos o parâmetro L , um array $k_n$ e o número de intervalos para a integração, conseguimos obter os coeficientes de Fourier pelos quais podemos criar aproximações de funções periódicas. Ao aplicarmos a função get_coefficients, onde por cada iteração de $k_n$ multiplicamos os valores do array f(x) ( função aplicada aos pontos desde 0 a L ) por $e^{\frac{-i x k}{L}}$ onde x é o array dos pontos em avalição, determinamos um array de valores complexos onde ao aplicarmos a regra de simpson no mesmo conseguimos chegar ao coeficiente pretendido para cada valor de $k_n$.


```python

def get_coefficients(func, L, ks, M):
    x = np.linspace(0, L, M + 1)
    coef = np.zeros(len(ks),dtype=complex)
    
    h = L/M
    for i, k in enumerate(ks):  # roda k_n pelo indice e respetivo valor
        f = func(x) * np.exp(-1.0j * x * k) /L
        coef[i] = h/3 * ( f[0] + f[-1] + 4 * f[1:-1:2].sum() + 2 * f[2:-2:2].sum() )  #regra de simpson 
    return coef

```

# Aproximação de funções com série finita

Numericamente não podemos somar uma série infinita, logo, temos limitar a série a um número máximo de valores de n. Com a série limitada a aproximação de ordem N de uma função será dada por, \begin{align}
f_{N}(x)=\sum_{n=-N}^{N}c_{n}e^{ik_{n}x}.
\end{align}

1. Considere a função periódica 
\begin{align}
f(x)=e^{-40(1-\sin^{2}(\pi x))}
\end{align}
de período $L=1$.

(a) Represente a função e as respectivas aproximações de Fourier para valores de N escolhidos entre potências de dois (\{2,4,8,16,32,64,128\}). Sugestão: Como a aproximação de Fourier poderá ter parte imaginária represente em dois gráficos diferentes a parte real e parte imaginária - faça isto sempre que for necessário.

#Responda Aqui
- Ao longo do aumento do valor da aproximação ( 2 a 128 ) conseguimos verificar uma aproximação à função como era pretendido. Conseguimos também verificar que com 16 aproximações o gráfico obtido ja coincide com o da função original, logo cálculos para aproximações superiores ao valor 16 já seriam desnecessários para esta função em específico.


- Os valores imaginários têm escala inferior à escala do gráfico da aproximação, logo ao serem representados em conjunto, apenas é visível uma linha. Ao serem representados num gráfico individual, conseguimos ver que os mesmos podem ser conciderados como sendo resíduos máquina, não apresentam ordem nem seguem uma configuração.

```python
# Responda Aqui

def f(x):
    return np.exp(-40 * (1 - np.sin(np.pi*x)**2))

def build_approx(ks, coef, x):
    fx = np.zeros(len(x), dtype=complex)   
    for i, k in enumerate(ks):
        fx += coef[i] * np.exp(1.0j * k * x)
    return fx

for i in range(1,8):
    ks = np.arange(-2**i,2**i+1) * 2 * np.pi/L
    coef = get_coefficients(f, L, ks, M)
    ff = build_approx(ks,coef, x )
    plt.plot(x,f(x),"k")
    plt.title("Gráficos de aproximações")
    plt.plot(x,ff.real,"r--",label= "Aproximação de valor {}".format(2**i))
    plt.legend()
    plt.show()
    plt.plot(x,ff.imag,color="#800009",label="valores imaginários para {} aproximações".format(2**i))
    plt.legend()
    plt.show()
    
```

(b) Para discutir a qualidade da aproximação podemos medir o desvio entre a função e a aproximação. Para este efeito, considere a grandeza,
\begin{align}
D_{N} &= \frac{\int_{0}^{L}dx\sqrt{\left|f(x)-f_{N}(x)\right|^{2}}}{\int_{0}^{L}dxf(x)}.
\end{align}
Represente os desvios $D_{N}$ para números inteiros consecutivos desde $0\le N<128$. Discuta o que observa e interprete os resultados.

#Responda Aqui

- Através do gráfico obtido do desvio padrão conseguimos verificar que com o aumento da aproximação demonstra-nos uma diminuição no valor do desvio padrão, como era de se esperar. Por outras palavras quanto mais pequeno for o valor do desvio padrão, mais os valores obtidos da aproximação tendem para a linha da função inicial. Conseguimos também ver que para o valor n = 16 temos um desvio padrão de $\cong$ 0.0025 que é bastante próximo de 0, não será então necessário fazer mais aproximações porque este valor já é satisfatório.

```python
# Responda Aqui


x2=np.arange(129)

denominador=get_coefficients(f, L, np.array([0]), M)

def simpson(x):  #recebe um array
    h=L/M
    return h/3 * (x[0] + x[-1] + 4*x[1:-1:2].sum() + 2 * x[2:-2:2].sum())

Dn = np.zeros(len(x2),dtype=np.float64)

for i in x2:
    ks = np.arange(-i,i+1) * 2 * np.pi/L
    coef = get_coefficients(f, L, ks,M)
    ff = build_approx(ks,coef, x)
    valor_int = np.sqrt((f(x)-ff)**2)
    
    numerador = simpson(valor_int)
    
    Dn[i]= (numerador/denominador).real
    
plt.plot(x2,Dn,".--")
plt.title("Desvio em função da aproximação")
plt.show()
plt.plot(x2,Dn,"r-.")
plt.yscale("log")
plt.title("log do desvio em função da aproximação")
plt.show()
```

(c) Em 1799 Parseval apresentou o teorema, 
\begin{align}
\sum_{n\in\mathbb{Z}}b_{n}^{*}a_{n}=\frac{1}{L}\int_{0}^{L}dxB^{*}(x)
\end{align}
$A(x)$onde $a_{n}$ e $b_{n}$ são respectivamente os coeficientes da decomposição de Fourier das funções $B(x)$ e $A(x)$. Uma aplicação possível é ver que a norma das duas representações deste objecto é conservada, 
\begin{align}
\sum_{n\in Z}\left|c_{n}\right|^{2}=\frac{1}{L}\int_{0}^{L}dx\left|f(x)\right|^{2}
\end{align}
quando limitamos a série a um número de termos finitos, podemos aferir a componente que falta calculando a diferença relativa,
\begin{align}
P_{M}=\left|1-\frac{\sum_{n=-M}^{M}\left|c_{n}\right|^{2}}{\frac{1}{L}\int_{0}^{L}dx\left|f(x)\right|^{2}}\right|.
\end{align}
Represente $P_{M}$ como função de $M$ e comente os resultados.

#Responda Aqui

- Através do gráfico obtido, conseguimos ver que que $P_M$ tende para 0 quando o valor de M tende a aumentar, que como valor de referência podemos tomar o 16 que como nas alineas anteriores é um bom valor para o qual não são necessárias mais aproximações. Este parâmetro que vem do teorema de Parseval é uma medida de precisão da aproximação de Fourier em relação à função referência, representa a diferença relativa entre as duas. 

```python
# Responda Aqui

pm = np.zeros(len(x2),dtype=np.float64)

def f2(x):
    return f(x)**2

denominador=get_coefficients(f2, L, np.array([0]), M)

def PM(x):
    return (x/denominador)

for i in x2:
    ks = np.arange(-i,i+1) * 2 * np.pi/L
    coef = get_coefficients(f, L, ks,M)
    numerador=(coef**2).sum()
    pm[i] = np.abs(1-PM(numerador).real)
    
plt.plot(x2,pm,color="#800009",linestyle=":",label = "valores de Pm")
plt.legend()
plt.show()
```

Através do teorema de Parseval, é possível ver que existem erros significativos durante a  aproximação de Fourier, em pontos de descontinuidade de funções.

# Propagação de ondas numa corda

Considere uma corda fixa pelas extremidades. A equação de movimento da altura em cada posição ($\psi(x,t)$) é dada por
\begin{align}
\frac{\partial^{2}\psi(x,t)}{\partial t^{2}}=c^{2}\frac{\partial^{2}\psi(x,t)}{\partial^{2}x}
\end{align}
onde $c$ é a com a condição fronteira determinada por $\psi(0,t)=\psi(L,t)=0$. 

1. Adimensionalize esta equação de modo a ter todas as quantidades físicas como quantidades adimensionais. 

#Responda Aqui

Para adimensionalizar a equação seguimos o seguinte pensamento: 

$$\frac{\partial^2\psi}{\partial t^2} = c^2 . \frac{\partial^2 \psi}{\partial x^2} $$

onde consideramos $ t\;' = \frac{t}{T} ; x\;' = \frac{x}{X} $


$$\frac{\partial}{\partial x } = \frac{\partial x^{'}\;}{\partial x} \frac{\partial}{\partial x^{'}\;} = \frac{1}{X} \frac{\partial}{\partial x^{'}\;} = \frac{\partial^2}{\partial x^2} = \frac{\partial}{\partial x } \left[\frac{1}{X} \frac{\partial}{\partial x^{'}\;}\right] = \frac{1}{X^2} \frac{\partial^2}{\partial x^{'2}\;}$$



Analogamente: 

$$\frac{\partial}{\partial t} = \frac{1}{T} \frac{\partial}{\partial t^{'}\;} ; \frac{\partial^2}{\partial t^2} = \frac{1}{T^2} \frac{\partial ^2}{\partial t^{'2}\;}$$


$\rightarrow$ Substituindo na eq. diferencial 

$$\frac{1}{T^2} \frac{\partial ^2 \psi}{\partial t^{'2}\;} = \frac{c^2}{X^2} \frac{\partial \psi^2}{\partial x^{'2}\;} \equiv \frac{\partial ^2 \psi}{\partial t^{'2}\; } = \frac{c^2 T^2}{X^2} \frac{\partial ^2 \psi}{\partial x^{'2}\;}$$

Aqui já temos os 3 membros da equação adimens.

T $\rightarrow$ tempo característico do sistema

X $\rightarrow$ comprimento característico do sistema

$\rightarrow$ o fator $c^2$ na equação de onda sugere que a onde se propaga à velidade c, pelo que podemos definir T como T  = $\frac{L}{c}$
