# 1. Série de Fourier

Uma função diz-se periódica se para todos os valores de $x$ verifica-se que $f(x+L)=f(x)$. Estas funções podem ser expressas à custa de uma base de funções periódicas, 
\begin{align}
f(x)=\sum_{n=-\infty}^{+\infty}c_{n}e^{ik_{n}x}
\end{align}
com os valores $k_{n}=2\pi n/L$ e os coeficientes $c_{n}$ dados por 
\begin{align}
c_{n}=\frac{1}{L}\int_{0}^{L}dxf(x)e^{-ik_{n}x}.
\end{align}
Esta expansão chama-se uma série de Fourier e pode ser entendida como decomposição numa base infinita de funções designadas por,
\begin{align}
\phi_{n}(x)=e^{ik_{n}x}
\end{align}
e na qual podemos definir um produto interno entre duas funções $f(x)$ e $g(x)$ como,
\begin{align}
f\cdot g=\frac{1}{L}\int_{0}^{L}dxf^{*}(x)g(x).
\end{align}
Com esta interpretação os coeficientes $c_{n}$ podem ser compreendidos como as projecções de uma função em cada elemento da base. 

1. Mostre analiticamente a ortogonalidade dos elementos da base, ou seja $\phi_{n}\cdot\phi_{m}=\delta_{mn}$.

#Responda Aqui


Sejam 

$\phi_n = e^{-iK_n x}, \space \space \space K_n = \frac{2\pi n}{L}$



$\phi_m = e^{iK_m x}, \space \space \space K_m = \frac{2\pi m}{L}$


Dois elementos da base. Fazemos então o produto das duas, obtendo:

$\phi_n \cdot \phi_m = \frac{1}{L}  \int_0^L  e^{ix(K_m - K_n)} dx =  \frac{1}{L} \int_0^L e^{\frac{2\pi}{L}ix(m-n)} dx $

Integra-se agora de 0 a L (um período) e prova-se a ortogonalidade entre dois elementos diferentes:

$ \frac{1}{L} \int_0^L e^{\frac{2\pi x i(m-n)}{L}} dx = \frac{e^{2\pi i (m-n)} - 1}{2\pi i(m-n)} = \frac{1-1}{2\pi i(m-n)} = 0, \ \ \ \ m \neq n$
$$$$
$ \frac{1}{L} \int_0^L e^{\frac{2\pi x i(m-n)}{L}} dx = \frac{1}{L} \int_0^Ldx = \frac{L}{L} = 1, \ \ m = n$

Aqui, vemos que o resultado da exponencial complexa, após a integração, é 1 $\forall$ $m \neq n$, uma vez que $m - n$ resultará em um inteiro e teremos um múltiplo de $2\pi$ no expoente. O resultado final da integral será então $0$, sempre que a condição $m\neq n$ for satisfeita. Caso $m = n$, a exponencial complexa resulta na constante 1  logo antes da integração, fazendo com que o resultado após a integração, de 0 a 1, resulte em $\frac{L}{L} = 1$ 
$$$$
Por fim, podemos escrever o resultado final como:
$$$$
$\phi_n \cdot \phi_m  = \frac{1}{L} \int_0^L e^{\frac{2\pi i x(m-n)}{L}} dx = δ_{mn}$

3. Construa uma função que calcule uma série de coeficientes de Fourier para uma função arbitrária usando o método de Simpson. A função deve receber como argumentos pelo menos o parâmetro $L$, um array de $k_{n}$ e o número de intervalos de integração. 

#Responda Aqui

```python
# Responda Aqui


def Coef(L,Kn,n,funcao):

    Kn = 1/L * Kn      
    x = np.linspace(0, L, n+1) # cria array intervalo de integração com n+1 pontos e n intervalos
    integrandos_matrix = np.zeros((len(Kn), len(x)), dtype=np.complex128)  # cria uma matriz onde vou guardar a função avaliada em cada ponto nas colunas para o Kn de cada linha
    f = funcao(x) # aplico o intervalo de integração na função escolhida
    w = np.ones(n+1)  # array com os pesos para a integração de simpson 1/3
    w[1:-1:2] *= 4  # pesos ímpares 
    w[2:-1:2] *= 2 # pesos pares
    
    coef = np.zeros(Kn.size, dtype=np.complex128)  # array onde vou guardar os coeficientes calculados
    
    for i in range(len(Kn)):  # percorrer os índices do array de Kn's
        
        integrandos_matrix[i,:] = f * np.exp(-1j*Kn[i]*x)   # numa mesma linha, guardo a função avaliada em todos os pontos do intervalo de int. para um Kn específico
        
        coef[i] = (w @ integrandos_matrix[i,:]) * 1/(3*n) # aqui, calculo o coeficiente fazendo o produto interno entre a matriz de pesos e o array da linha i da matriz (que contem a função avaliada nos pontos pra cada Kn)
        
        # notar que L irá cortar (e também é igual a)


    return coef


    
    
```

# Aproximação de funções com série finita

Numericamente não podemos somar uma série infinita, logo, temos limitar a série a um número máximo de valores de n. Com a série limitada a aproximação de ordem N de uma função será dada por, \begin{align}
f_{N}(x)=\sum_{n=-N}^{N}c_{n}e^{ik_{n}x}.
\end{align}

O fenômeno de oscilações de Gibbs é um efeito que ocorre quando uma função apresenta descontinuidades e é aproximada por uma série de Fourier truncada. Quando a série de Fourier possui um número limitado de termos, a aproximação da função fica cada vez melhor quando aumenta-se o número de termos usados, mas em torno das descontinuidades da função original haverá sempre oscilações, um erro que estará presente devido à esta descontinuidade.


Na função $f(x)$ em questão, é possível observar estas oscilações em $x = 0$, $\ x = 0.5$ e em $x = 1$. Elas diminuem com o aumento de $N$, mas estão sempre presentes.

Diferente da função anterioremente analisada, que envolvia um exponencial (contínuo em todo ℜ) a função desta questão possui pontos de descontinuidade, o que leva ao fenômeno de oscilações de Gibbs em torno destes. A análise de $D_n$ e de $P_m$ torna ainda mais visível o quanto estes pontos de descontinuidade impactam na aproximação utilizando da série de Fourrier. Para o mesmo $N=20$, que antes permitia uma precisão de algo em torno de 3 a 5 casas decimais, agora tem-se uma precisão de 1 casa decimal.
