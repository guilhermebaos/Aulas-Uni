```python
from typing import Callable
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dft(yn: np.ndarray, real: bool=True) -> np.ndarray:
    """
        Devolve os `N = len(yn)` primeiros coeficientes da Transformada de Fourier Discreta da sequncia de nmeros `yn`. Se `real == True` ento assume que os nmeros so reais e usa isso para calcular apenas metade dos coeficientes.
    """

    # Determinar N
    N = len(yn)

    # Determinar os coeficientes
    if real:
        nn = np.arange(N)
        kk = np.arange(N//2 + 1)
        
        ck = np.zeros(N, dtype=np.complex128)
        ck[:(N//2) + 1] = np.exp(-2j * np.pi * np.outer(kk, nn) / N) @ yn

        for k in range(1, N//2 + 1):
            ck[N - k] = np.conjugate(ck[k])

        return ck
    
    else:
        nn = np.arange(N)
        kk = np.arange(N)
        
        return np.exp(-2j * np.pi * np.outer(kk, nn) / N) @ yn
```

## Alnea 1

```python
def f(x: np.ndarray) -> np.ndarray:
    dec, _ = np.modf(x)
    dec = 1 - np.floor(2 * dec)
    return dec


N = 1000
tt = np.linspace(0, 0.9999, N)
yy = f(tt)

ii = np.arange(0, N)
ck = dft(yy)

plt.plot(ii, np.abs(ck))
plt.xlim(-1, 100)
plt.grid()
plt.show()
```
