```python
from typing import Callable
import numpy as np
import matplotlib.pyplot as plt
```


## Alínea 1

```python
# Abrir os dados
dow = np.loadtxt("04-dow.txt", float)

# Mostrar os dados num gráfico
plt.plot(dow)
plt.title("Dow Jones Industrial Average (2006-2010)")
plt.xlabel("Dias desde 2006")
plt.ylabel("Pontos")

# Acho que dá para ver 2008 xD
```

## Alínea 2

```python
# Coeficientes
cc = np.fft.rfft(dow)
Nc = len(cc)
```

## Alínea 3

```python
# Reter apenas os primeiros 10% dos índices
cc10 = np.zeros_like(cc)
cc10[:Nc//10] = cc[:Nc//10]
```

## Alínea 4

```python
# Reconstruir os dados filtrados
dow10 = np.fft.irfft(cc10)

# Mostrar os dados num gráfico
plt.plot(dow, label="Original")
plt.plot(dow10, label="Alisados")
plt.legend()
plt.title("Dow Jones Industrial Average (2006-2010)")
plt.xlabel("Dias desde 2006")
plt.ylabel("Pontos")
```

## Alínea 5

```python
# Reter apenas os primeiros 2% dos índices
cc2 = np.zeros_like(cc)
cc2[:Nc//50] = cc[:Nc//50]

# Reconstruir os dados filtrados
dow2 = np.fft.irfft(cc2)

# Mostrar os dados num gráfico
plt.plot(dow, label="Original")
plt.plot(dow2, label="Alisados")
plt.legend()
plt.title("Dow Jones Industrial Average (2006-2010)")
plt.xlabel("Dias desde 2006")
plt.ylabel("Pontos")
```
