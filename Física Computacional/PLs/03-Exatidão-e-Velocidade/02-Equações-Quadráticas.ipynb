## Alínea 1

```python
def quadratica(a: float, b: float, c: float) -> tuple[float, float]:
    binomio = (b ** 2 - 4 * a * c) ** 0.5 / (2 * a)
    termo = -b / (2 * a)
    return termo + binomio, termo - binomio

print(quadratica(0.001, 1000, 0.001))
```

## Alínea 2

Começamos por deduzir a expressão:
$$
\begin{align*}

\frac{-b \pm \sqrt{b^2 - 4ac}}{2a} &= \frac{(-b \pm \sqrt{b^2 - 4ac})(-b \mp \sqrt{b^2 - 4ac})}{2a(-b \mp \sqrt{b^2 - 4ac})} =
\\\\

&= \frac{b^2 - b^2 + 4ac}{2a(-b \mp \sqrt{b^2 - 4ac})} =
\\\\

&= \frac{2c}{-b \mp \sqrt{b^2 - 4ac}}

\end{align*}
$$

Temos então:

```python
def quadratica_inversa(a: float, b: float, c: float) -> tuple[float, float]:
    binomio = (b ** 2 - 4 * a * c) ** 0.5
    return (2 * c) / (-b + binomio), (2 * c) / (-b - binomio)

print(quadratica_inversa(0.001, 1000, 0.001))
```

Os valores que obtivemos são muito diferentes dos valores acima. [Notamos](https://www.wolframalpha.com/input?i=0.001x%5E2+%2B+1000x+%2B+0.001+%3D+0) que as soluções são aproximadamente:

$$
x_1 \approx -1 \times 10^6 \quad\quad x_2 \approx -1 \times 10^{-6}
$$

Ou seja, a segunda solução é muito mais precisa. Isto deve-se ao facto de ocorrer [cancelação catastrófica](https://www.notion.so/guilhermebaos/3-Cancela-o-Catastr-fica-882e4528bff5489d999baf3071b6cd58?pvs=4).

---

Neste caso o efeito deve-se ao numerador ser um número muito pequeno!
