Vamos comear por calcular a derivada da funo $f(x) = x(x-1) = x^2 - x$:

$$

f'(x) = 2x - 1

$$

Temos ento $f'(1) = 1$.

## Alnea 1

```python
import numpy as np

# Type hint de funes
from typing import Callable


def f(x: float) -> float:
    return x * (x - 1)


# Calcular a derivada num ponto usando a definio
def derivada_por_def(func: Callable, x: float, h = 1e-02) -> float:
    return (func(x + h) - func(x)) / h


print(derivada_por_def(f, 1))
```

Podemos perceber o facto de o valor obtido diferir do valor real considerando a definio de derivada como o declive da reta tangete ao grfico da funo.

Neste caso, ao usar $\delta = 10^{-2}$ no estamos a calcular o declive da reta tangente em $x 1$ mas sim o declive de uma reta secante que interseta o grfico da funo em $x = 1$ e em $x = 1 + \delta$.

## Alnea 2

```python
hh = 10.0 ** np.arange(-2, -24, -2)

print(derivada_por_def(f, 1, hh))
```

Inicialmente a preciso melhora pois estamos a tomar uma melhor aproximao da derivada (uma reta sencante mais prxima da reta tangente), mas para $\delta$ suficientemente pequeno a aproximao piora pois a diferena que estamos a fazer no numerador tem um erro relativo muito grande, levando a [cancelao catastrfica](https://www.notion.so/guilhermebaos/3-Cancela-o-Catastr-fica-882e4528bff5489d999baf3071b6cd58?pvs=4).
