# Exercício 3



Vamos começar por calcular a derivada da função $f(x) = x(x-1) = x^2 - x$:

$$

f'(x) = 2x - 1

$$

Temos então $f'(1) = 1$.

## Alínea 1

```python
import numpy as np

# Type hint de funções
from typing import Callable


def f(x: float) -> float:
    return x * (x - 1)


# Calcular a derivada num ponto usando a definição
def derivada_por_def(func: Callable, x: float, h = 1e-02) -> float:
    return (func(x + h) - func(x)) / h


print(derivada_por_def(f, 1))
```

Podemos perceber o facto de o valor obtido diferir do valor real considerando a definição de derivada como o declive da reta tangete ao gráfico da função.

Neste caso, ao usar $\delta = 10^{-2}$ não estamos a calcular o declive da reta tangente em $x 1$ mas sim o declive de uma reta secante que interseta o gráfico da função em $x = 1$ e em $x = 1 + \delta$.

## Alínea 2

```python
hh = 10.0 ** np.arange(-2, -24, -2)

print(derivada_por_def(f, 1, hh))
```

Inicialmente a precisão melhora pois estamos a tomar uma melhor aproximação da derivada (uma reta sencante mais próxima da reta tangente), mas para $\delta$ suficientemente pequeno a aproximação piora pois a diferença que estamos a fazer no numerador tem um erro relativo muito grande, levando a [cancelação catastrófica](https://www.notion.so/guilhermebaos/3-Cancela-o-Catastr-fica-882e4528bff5489d999baf3071b6cd58?pvs=4).
