```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definir funções
def fatorial(n: int) -> int:
    return np.arange(1, n + 1).prod()


def fatorial_float(n: float) -> float:
    return np.arange(1.0, n + 1, 1.0).prod()
```

```python
# Executar as funções e medir o tempo
N = 16
nn = 2 ** np.arange(1, N)

tf = np.zeros(N - 1)
tff = np.zeros(N - 1)
for i, n in enumerate(nn):
    tempof = %timeit -n 100 -r 50 -o fatorial(n)
    tempoff = %timeit -n 100 -r 50 -o fatorial_float(n)

    tf[i] = tempof.average
    tff[i] = tempoff.average

plt.plot(nn, tf, "o--", label="Fatorial inteiro")
plt.plot(nn, tff, "o--", label="Fatorial float")
plt.xlabel("Valor de x")
plt.ylabel("Tempo em segundos")
plt.legend()
plt.show()
```

```python
# Fatorial com inteiros em Python
def fatorial_int(n: int) -> int:
    total = 1
    for i in range(1, n + 1):
        total *= i
    return total

print(fatorial(200))
print(fatorial_int(200))
print(fatorial_float(200))
```
