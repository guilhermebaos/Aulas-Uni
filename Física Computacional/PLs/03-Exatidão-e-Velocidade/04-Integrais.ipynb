## Alínea 1

```python
import numpy as np
from typing import Callable


# Integral de Riemann
def integral_riemann(func: Callable, start: float, end: float, N=100):
    h = (end - start) / N
    return sum(h * func(x) for x in np.linspace(start, end, N + 1)[:-1])


# Função em estudo
def f(x: float) -> float:
    return np.sqrt(1 - x**2)


print(np.pi / 2)
print(integral_riemann(f, -1, 1))
```

## Alínea 2

```python
nn = 10 ** np.arange(2, 8, 2)

print(np.pi / 2)

for item in nn:
    print(integral_riemann(f, -1, 1, item))
```
