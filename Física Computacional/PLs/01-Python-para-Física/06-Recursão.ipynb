## Alnea 1

```python
# Parmetros
N = 100


# Funo para calcular um nmero de Catalan recursivamente
def catalan(n: int) -> int:
    if n == 0:
        return 1
    else:
        return int((4 * n - 2) / (n + 1) * catalan(n - 1))


for i in range(0, 100 + 1):
    print(f"O {i} nmero de Catalan  {catalan(i)}")
```

## Alnea 2

```python
# Parmetros
N = 108
M = 192


# Mximo divisor comum
def gcd(n: int, m: int) -> int:
    # Certificar que n > m
    if n < m:
        n, m = m, n
    
    if m == 0:
        return n
    else:
        # Fazer gcd(m, n - km) onde k  um inteiro
        return gcd(m, n % m)


print(f"O menor mltiplo comum entre {N} e {M}  {gcd(N, M)}")
```
