## Alínea 1

```python
# Parâmetros
N = 100


# Função para calcular um número de Catalan recursivamente
def catalan(n: int) -> int:
    if n == 0:
        return 1
    else:
        return int((4 * n - 2) / (n + 1) * catalan(n - 1))


for i in range(0, 100 + 1):
    print(f"O {i}º número de Catalan é {catalan(i)}")
```

## Alínea 2

```python
# Parâmetros
N = 108
M = 192


# Máximo divisor comum
def gcd(n: int, m: int) -> int:
    # Certificar que n > m
    if n < m:
        n, m = m, n
    
    if m == 0:
        return n
    else:
        # Fazer gcd(m, n - km) onde k é um inteiro
        return gcd(m, n % m)


print(f"O menor múltiplo comum entre {N} e {M} é {gcd(N, M)}")
```
