## Alínea 1

```python
# Começamos por definir a função prod (produtório) de forma semelhante a range
def prod(start: int, end: int, step=1) -> int:
    total = 1
    for i in range(start, end, step):
        total *= i
    return total


# Definimos a função que calcula os coeficientes binomiais
def binomial(n: int, k: int) -> int:
    if k == 0 or k == n:
        return 1
    else:
        return prod(n - k + 1, n + 1) // prod(1, k + 1)
```

## Alínea 2

```python
# Parâmetro
num = 20

# Determinar as linhas do triângulo
max_num = binomial(num, num // 2)
len_num = len(str(max_num))

linhas = []
for n in range(0, num):
    linhas += [""]
    for k in range(0, n + 1):
        linhas[-1] += f"{binomial(n, k):^{len_num}}"

max_len = len(linhas[-1])
for item in linhas:
    print(f"{item:^{max_len}}")
```

# Alínea 3

```python
# Parâmetros
N = 100
C = 50


# Probabilidade de a moeda, ao ser lançada {lances} vezes sair {cara} vezes cara
def prob_cara(lances: int, cara: int) -> float:
    return binomial(lances, cara) / (2 ** lances)



# Probabilidade de a moeda, ao ser lançada {lances} vezes sair {cara} ou mais vezes cara
def prob_cara_ou_mais(lances: int, cara: int) -> float:
    total = 0
    for k in range(cara, lances + 1):
        total += binomial(lances, k)
    return total / (2 ** lances)


# Caso pedido
print(f"A probabilidade de ao lançar {N} vezes a moeda sair {C} vezes cara é {prob_cara(N, C):.4f}")
print(f"A probabilidade de ao lançar {N} vezes a moeda sair {C} ou mais vezes cara é {prob_cara_ou_mais(N, C):.4f}")
```
