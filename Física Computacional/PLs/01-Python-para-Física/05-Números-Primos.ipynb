```python
# Parmetros
N = 10_000


# Calcular primos
primos = [2]
for n in range(3, N + 1):
    for p in primos:

        # Encontramos um divisor de n
        if n % p == 0:
            break
            
        # No encontramos um divisor de n e podemos parar de procurar
        elif p > n ** 0.5:
            primos += [n]
            break
print("\n".join(map(str, primos)))
```
