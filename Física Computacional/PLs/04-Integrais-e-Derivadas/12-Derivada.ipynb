```python
from typing import Callable
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Conjunto de funes para calcular derivadas de primeira ordem
from typing import Callable

def forward(func: Callable, h: float = 1e-6) -> Callable:
    """
        Calcula a derivada de `func` usando a diferena progressiva com um passo de `h`.

        ### Retorno
        Devolve uma funo `fprime` que recebe como argumento `x` e devolve a derivada de `func` em `x`.
    """

    return lambda x: (func(x + h) - func(x)) / h


def backward(func: Callable, h: float = 1e-6) -> Callable:
    """
        Calcula a derivada de `func` usando a diferena regressiva com um passo de `h`.

        ### Retorno
        Devolve uma funo `fprime` que recebe como argumento `x` e devolve a derivada de `func` em `x`.
    """

    return lambda x: (func(x) - func(x-h)) / h


def central(func: Callable, h: float = 1e-4) -> Callable:
    """
        Calcula a derivada de `func` usando a diferena central com um passo de `h`.

        ### Retorno
        Devolve uma funo `fprime` que recebe como argumento `x` e devolve a derivada de `func` em `x`.
    """

    return lambda x: (func(x + h/2) - func(x - h/2)) / h
```

Temos que:

$$
\begin{align*}
f(x) &= 1 + \frac12\tanh(2x)
\\\\
f'(x) &= \text{sech}^2(2x)
\end{align*}
$$

```python
def func(x: (float | np.ndarray)) -> (float | np.ndarray):
    return 1 + np.tanh(2 * x) / 2


def funcprime(x: (float | np.ndarray)) -> (float | np.ndarray):
    return (1 / np.cosh(2 * x))**2


# Parmetro
h = 0.1

tt = np.linspace(-2, 2, 1000)
ff = funcprime(tt)
fff = forward(func, h)(tt)
ffb = backward(func, h)(tt)
ffc = central(func, h)(tt)


plt.plot(tt, ff, label="Derivada Exata")
plt.plot(tt, fff, label="Derivada Progressiva")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.legend(bbox_to_anchor=(1.05, 1), borderaxespad=0)
plt.show()


plt.plot(tt, ff, label="Derivada Exata")
plt.plot(tt, ffb, label="Derivada Regressiva")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.legend(bbox_to_anchor=(1.05, 1), borderaxespad=0)
plt.show()


plt.plot(tt, ff, label="Derivada Exata")
plt.plot(tt, ffc, label="Derivada Central")
plt.xlabel("x")
plt.ylabel("f'(x)")
plt.legend(bbox_to_anchor=(1.05, 1), borderaxespad=0)
plt.show()

```
