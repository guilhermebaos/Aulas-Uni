```python
import numpy as np
import matplotlib.pyplot as plt

from typing import Callable
```

## Alnea 1

```python
# Regra de Simpson para 2n passos
def simpson_pontos(start: float, end: float, func: Callable, n: float):
    # Abcissas
    h = (end - start) / (2 * n)
    tt = np.linspace(start, end, 2 * n + 1)
    vv = func(tt)

    # rea segmentos
    aa = vv[:-2:2] + 4 * vv[1:-1:2] + vv[2: :2]
    integral = np.concatenate((np.array([0]), aa.cumsum()))

    return (tt[: :2], h * (integral) / 3)


# Funo a integrar
def f(x: float) -> float:
    return np.e**(-x**2)


# Limites do integral
start = 0
end = 3

# Pontos do integral
steps = 30

# Intevalos por ponto do integral
intervalos = 100

xx, yy = simpson_pontos(start, end, f, (steps - 1) * intervalos // 2)
print(yy[-1])
```
