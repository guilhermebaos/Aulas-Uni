```python
import numpy as np
import matplotlib.pyplot as plt
```

## Alnea 1

```python
tt, vv = np.loadtxt("01-velocities.txt", unpack=True)

# Executar a regra do trapzio para uma lista discreta
def trapezio_lst(h, vv):
    # Passo e resultado do integral
    n = len(vv)
    xx = np.zeros(n)

    # rea dos trapzios
    aa = h * (vv[:-1] + vv[1:]) / 2

    # Integral at cada ponto
    for i in range(n):
        xx[i] = aa[:i].sum()

    return xx

xx = trapezio_lst(tt[1] - tt[0], abs(vv))
print(xx[-1])
```

## Alnea 2

```python
plt.plot(tt, xx, label="Posio")
plt.plot(tt, vv*10, label="Velocidade")
plt.legend()
plt.show()
```
