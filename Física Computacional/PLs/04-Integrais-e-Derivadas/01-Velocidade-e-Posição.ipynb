```python
import numpy as np
import matplotlib.pyplot as plt
```

## Alínea 1

```python
tt, vv = np.loadtxt("01-velocities.txt", unpack=True)

# Executar a regra do trapézio para uma lista discreta
def trapezio_lst(h, vv):
    # Passo e resultado do integral
    n = len(vv)
    xx = np.zeros(n)

    # Área dos trapézios
    aa = h * (vv[:-1] + vv[1:]) / 2

    # Integral até cada ponto
    for i in range(n):
        xx[i] = aa[:i].sum()

    return xx

xx = trapezio_lst(tt[1] - tt[0], abs(vv))
print(xx[-1])
```

## Alínea 2

```python
plt.plot(tt, xx, label="Posição")
plt.plot(tt, vv*10, label="Velocidade")
plt.legend()
plt.show()
```
