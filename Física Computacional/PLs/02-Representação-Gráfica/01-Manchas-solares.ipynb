# Exerccio 1



```python
import numpy as np
import matplotlib.pyplot as plt
```

## Alnea 1

```python
# Ler o valores
with open("01-sunspots.txt") as file:
    dados = file.read().split("\n")

# Converter os dados para pares de items float
dados = list(map(lambda x: list(map(float, x.split("\t"))), dados))

# Separar em tempo e nmero de manchas solares
tt = np.array([x[0] for x in dados])
nn = np.array([x[1] for x in dados])

# Representar graficamente
plt.plot(tt, nn)
plt.xlabel("Meses desde janeiro de 1749")
plt.ylabel("Nmero de manchas solares")
plt.show()
```

## Alnea 2

```python
# Limitar aos primeiros 1000 pontos
tt = tt[:1000]
nn = nn[:1000]

# Representar graficamente
plt.plot(tt, nn)
plt.xlabel("Meses desde janeiro de 1749")
plt.ylabel("Nmero de manchas solares")
plt.show()
```

## Alnea 3

```python
# Definir a mdia mvel de um array
def media_movel(r: int, arr: np.ndarray) -> np.ndarray:
    media = []
    # Fazer a mdia com os valores disponveis

    
    # Esta forma  pouco eficiente mas funciona bem para todos os valores,
    # incluindo o incio e o fim dos dados
    for i in range(0, len(arr)):
        start = max(0, i-r)
        end = min(len(arr), i+r+1)
        valores = end - start
        media += [sum(arr[start:end])/valores]
    media = np.array(media)
    return media


teste = np.array([1, 2, 3, 3, 2, 1])
print(teste)
print(media_movel(1, teste))

# Fazer a mdia mvel do nmero de manchas solares
nm = media_movel(5, nn)

# Representar graficamente
plt.plot(tt, nn, label="Manchas solares")
plt.plot(tt, nm, label="Mdia mvel de manchas solares")
plt.xlabel("Meses desde janeiro de 1749")
plt.ylabel("Nmero de manchas solares")

plt.legend()
plt.show()
```
