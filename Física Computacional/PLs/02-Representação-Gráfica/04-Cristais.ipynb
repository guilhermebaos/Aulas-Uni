```python
import numpy as np
import matplotlib.pylab as plt
```

## Alnea 1

```python
# Parmetro
L = 3

# Posies de grelha cbica
xx, yy, zz = np.mgrid[0:L, 0:L, 0:L]

# Cores
cores = 1 - 2 * ((xx + yy + zz) % 2)

# Fazer o Grfico
eixos = plt.axes(projection="3d")
eixos.set_axis_off()
eixos.scatter(xx, yy, zz, s=100, c=cores, cmap="Paired")

plt.show()
```

## Alnea 2

```python
# Parmetro
L = 2

# Posies de grelha cbica
xx, yy, zz = np.mgrid[-1:L, -1:L, -1:L]

ss = (xx + yy + zz) % 2 * 100


# Fazer o Grfico
eixos = plt.axes(projection="3d")
eixos.set_axis_off()
eixos.scatter(xx, yy, zz, c="orange", s=ss)

plt.show()
```
