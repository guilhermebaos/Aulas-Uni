# Exercício 7




Vamos definir algumas operações úteis com números complexos na forma algébrica:

$$
z = a + bi
$$

Temos que:

$$
z^2 = (a^2 - b^2) + 2abi
$$

```python
import numpy as np
import matplotlib.pylab as plt
```

```python
class Complexo():
    def __init__(self, a: float, b: float) -> None:
        self.a = a
        self.b = b
    

    def quadrado(self):
        self.a, self.b = self.a ** 2 - self.b ** 2, 2 * self.a * self.b
        return self
    

    def norma(self):
        return (self.a**2 + self.b**2)**0.5


    def __str__(self):
        return f"{self.a} + {self.b}i"


    def __add__(self, other):
        return Complexo(self.a + other.a, self.b + other.b)
```

```python
def mandelbrot(xmin: float, xmax: float, ymin: float, ymax: float, GRELHA=100, MAX_ITER=100) -> None:
    # Grelha
    cc = np.zeros((GRELHA, GRELHA), dtype="object")

    for y, b in enumerate(np.linspace(ymin, ymax, GRELHA)):
        for x, a in enumerate(np.linspace(xmin, xmax, GRELHA)):
            cc[y][x] = Complexo(a, b)

    # Ver os pontos do conjunto de Mandelbrot
    ii = np.zeros((GRELHA, GRELHA), dtype=int)

    for y, row in enumerate(cc):
        for x, c in enumerate(row):
            z = Complexo(0, 0)
            for iter in range(MAX_ITER):
                z = z.quadrado() + c

                if z.norma() > 2:
                    ii[y][x] = iter
                    break
            else:
                ii[y][x] = -1


    # Mostrar a imagem
    plt.imshow(ii, origin="lower", extent=[xmin, xmax, ymin, ymax])
    plt.hot()
    plt.show()

# Ver o conjunto de Mandelbrot e fazer zoom numa parte
mandelbrot(-2, 2, -2, 2)
mandelbrot(-1, -0.5, 0, 0.5, GRELHA=200)
mandelbrot(-1, -0.9, 0.25, 0.35, GRELHA=300)
```
