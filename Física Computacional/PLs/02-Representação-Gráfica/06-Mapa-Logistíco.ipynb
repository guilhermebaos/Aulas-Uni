```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parâmetros
ITER = 1000
START = 0
END = 4
PONTOS = 4001


# Mapa Logistíco
def ml(r: float, x: float) -> float:
    return r * x * (1 - x)


# Calcular o mapa logistíco para vários r
rr = np.linspace(START, END, PONTOS)
xx = np.full((1, PONTOS), 0.5).flatten()

for _ in range(ITER):
    xx = ml(rr, xx)

for _ in range(ITER):
    xx = ml(rr, xx)
    plt.scatter(rr, xx, color="blue", alpha=1/ITER * 2, s=0.3)

plt.show()

```
