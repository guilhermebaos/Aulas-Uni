```python
import matplotlib.pyplot as plt
import numpy as np
```

## Alínea 1

```python
# Valores de theta
tt = np.linspace(0, 2 * np.pi, 100)

# Pontos no gráfico
xx = 2 * np.cos(tt) + np.cos(2 * tt)
yy = 2 * np.sin(tt) - np.sin(2 * tt)

# Gráfico
plt.plot(xx, yy)
plt.show()
```

## Alínea 2

```python
def plot_polar(f, start=0, end=10 * np.pi, points=1000):
    # Valores de theta
    tt = np.linspace(start, end, points)
    rr = f(tt)

    # Pontos no gráfico
    xx = rr * np.cos(tt)
    yy = rr * np.sin(tt)

    # Gráfico
    plt.plot(xx, yy)
    plt.show()


def espiral(theta: float) -> float:
    return theta**2


plot_polar(espiral)
```

```python
def fey(theta: float) -> float:
    return np.exp(np.cos(theta)) - 2 * np.cos(4 * theta) + np.sin(theta/12) ** 5


plot_polar(fey, end=24 * np.pi, points=3000)
```
