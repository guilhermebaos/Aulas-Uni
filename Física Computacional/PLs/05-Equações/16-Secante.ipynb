```python
G = 6.674e-11
M = 5.974e24
m = 7.348e22
R = 3.844e8
omega = 2.662e-06

def f(r: float) -> float:
    return (G * M) / r**2 - (G * m) / (R - r)**2 - omega**2 * r


xx = np.linspace(1, 100, 1001)
yy = f(xx)

plt.plot(xx, yy)
plt.plot(xx, xx * 0)
```

```python
def secante(func: Callable, x0: float, x1: float, eps: float = 10e-6, maxI: int = 1000) -> float:
    """
        Resolve a equao `func(x) = 0` usando o mtodo da secante comeando com estimativas `x0` e `x1` e parando quando o erro for menor que `eps fazendo no mximo `maxI` iteraes.

        ### Argumentos
        func: Uma funo que aceita um float
        x0: Primeira estimativa.
        x1: Segunda estimativa.
        eps: Erro mximo permitido.
        maxI: Nmero mximo de iteraes, se for ultrapassado o mtodo levanta uma exceo.

        ### Retorno
        x: O valor x que  soluo da equao.
    """
    
    # Incio do algoritmo
    i = 0
    erro = eps + 1
    while abs(erro) > eps:
        # Dar um passo
        x2 = x1 - func(x1) * (x1 - x0)/(func(x1) - func(x0))

        # Calcular o erro
        erro = abs(x2 - x1)
        
        # Prximo passo
        x0 = x1
        x1 = x2

        # Nmero mximo de iteraes
        i += 1
        if i > maxI:
            raise ValueError(f"No foi possvel encontrar a raz da funo com a preciso desejada com menos de maxI = {maxI} iteraes!")
    
    return x1
```

```python
l = secante(f, 3e+08, 2.8e8)
print(f"O ponto de Lagrange est em r = {l:e}.")
```
