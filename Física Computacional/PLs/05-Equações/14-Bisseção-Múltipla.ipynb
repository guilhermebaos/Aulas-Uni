```python
from typing import Callable
import numpy as np
import matplotlib.pyplot as plt
```

## Alínea 1

```python
# Parâmetros (SI)
m = 9.1094e-31
q = 1.602e-19
V = 20 * q
hbar = 1.055e-34
w = 1e-09

# Lado esquerdo da equação
def lhs(x: float) -> float:
    return np.tan(w**2 * m * x / (2 * hbar**2))


# Lado direitao da equação
def rhs(x: float, n: int) -> float:
    # Caso ímpar
    if n % 2 == 1:
        return -np.sqrt(x / (V - x))
    # Caso par
    else:
        return np.sqrt((V - x) / x)


xx = np.linspace(0.001 * q, 3 * q, 10001)
yy1 = lhs(xx)
yy2 = rhs(xx, 0)
yy3 = rhs(xx, 1)

plt.plot(xx, yy1, label="yy1")
plt.plot(xx, yy2, label="yy2")
plt.plot(xx, yy3, label="yy3")

axis = plt.gca()
axis.set_ylim([-10, 10])

plt.legend()
plt.show()
```
