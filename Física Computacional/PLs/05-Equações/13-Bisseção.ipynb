```python
def bissect(func: Callable, a: float = 0, b: float = 1, eps: float = 10e-6) -> float:
    """
        Resolve a equao `func(x) = 0` usando o mtodo da bisseo no intervalo [a, b] parando quando o erro for menor que `eps`.

        ### Retorno
        O valor x que  soluo da equao.
    """

    erro = eps + 1
    while abs(erro) > eps:
        c = (a+b)/2

        fa, fb, fc = func(a), func(b), func(c)

        # Verificar se algum dos valores  um zero
        if fa == 0:
            return a
        elif fb == 0:
            return b
        elif fc == 0:
            return c

        # Troca de sinal em [a, c]
        elif fa * fc < 0:
            b = c
        
        # Troca de sinal em [c, b]
        elif fc * fb < 0:
            a = c
        
        erro = (b - a)/2
    
    return (a+b)/2
```

```python
def f(x: float) -> float:
    return 5 * np.exp(-x) + x - 5


# Encontrar a zona onde esto os zeros
xx = np.linspace(0, 100, 1001)
yy = f(xx)

plt.plot(xx, yy)
```

```python
# Encontrar o zero que no  x = 0
sol = bissect(f, 1e-6, 10)
```

```python
# Determinar a constante de Wien
h = 6.626e-34
c = 299_792_458
k = 1.38e-23

B = (h * c) / (k * sol)
print(f"A constante de Wien  {B:.6f}mK")
```
