## Modelo

Vamos começar por construir um modelo:
- $A$: Quantidade em metros produzida do tecido A
- $B$: Quantidade em metros produzida do tecido B

Temos então o seguinte problema de otimização linear:
$$
\begin{alignat*}{2}

\max_x &\quad& &500A + 200B
\\\\

\text{s.t}& &&0.4A + 0.5B \le 100
\\
&&&0.5A + 0.2B \le 100
\\

&&&0.3A + 0.8B \le 120
\\
&&& A, B \ge 0

\end{alignat*}
$$

## Código

```python
from pyscipopt import Model

# Criar o modelo
m = Model()

# Criar variáveis
A = m.addVar("Tecido A")
B = m.addVar("Tecido B")

# Adicionar restrições (constraints)
m.addCons(0.4 * A + 0.5 * B <= 100)
m.addCons(0.5 * A + 0.2 * B <= 100)
m.addCons(0.3 * A + 0.8 * B <= 120)

m.addCons(A >= 0)
m.addCons(B >= 0)

# Adicionar objetivo
m.setObjective(500*A + 200*B, "maximize")

# Otimizar o modelo (encontrar a solução ótima)
m.optimize()

# Obtemos todas as variáveis usando m.getVars() e depois, para cada variável i mostramos o seu nome (i.name) e o seu valor (m.getVal(i))
print([(i.name, m.getVal(i)) for i in m.getVars()])
```

A solução ótima é então:
- Produzir 176.47 metros do tecido A
- Produzir 58.82 metros do tecido B
